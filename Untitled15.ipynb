```python
from google.colab import files
import heapq

uploaded = files.upload()
filename = list(uploaded.keys())[0]
city = {}
ht = {}


with open(filename, "r") as file:
    for line in file:
        arr = line.split()
        for i in range(2, len(arr), 2):
            if arr[0] not in city:
                city[arr[0]] = {}
            city[arr[0]][arr[i]] = int(arr[i+1])
        ht[arr[0]] = int(arr[1])

def astarsearch(graph, huit, start, end):
    heap = []
    heapq.heappush(heap, (huit[start], 0, start, []))
    visited = set()

    while heap:
        f, g, current, path = heapq.heappop(heap)
        if current in visited:
            continue
        visited.add(current)
        path = path + [current]

        if current == end:
            return path, g

        for nei, cost in graph.get(current, {}).items():
            if nei not in visited:
                g1 = g + cost
                f1 = g1 + huit[nei]
                heapq.heappush(heap, (f1, g1, nei, path))

    return None, float('inf')


start = "Arad"
goal = "Bucharest"
print(f"Start node: {start}")
print(f"Destination: {goal}")

path, cost = astarsearch(city, ht, start, goal)

if path:
    path_str = " -> ".join(path)
    print(f"Path: {path_str}")
    print(f"Total distance: {cost} km")
else:
    print("NO PATH FOUND")
```

```python
#first imput
import random
capital = 1000
price = [-1.2, 3.4, -0.8, 2.1, -2.5, 1.7, -0.3, 5.8, -1.1, 3.5]
population = [{"stop_loss": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "take_profit": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "trade_size": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)])},
 {"stop_loss": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "take_profit": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "trade_size": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)])},
  {"stop_loss": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "take_profit": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "trade_size": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)])},
   {"stop_loss": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "take_profit": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)]), "trade_size": random.choice([random.randint(1, 99), round(random.uniform(1, 9.9), 1)])}]
generations = 10
final_profit = {}
states = {}

def fitness(cell, capital, price):
  for i in range(10):
    price_change = price[i]
    trade_size = (cell["trade_size"]*capital)/100
    if price_change < 0:
      if abs((trade_size*price_change)/100)<abs((trade_size*cell["stop_loss"])/100):
        profit_loss = (trade_size*price_change)/100
      elif abs((trade_size*price_change)/100)>abs((trade_size*cell["stop_loss"])/100):
        profit_loss = -((trade_size*cell["stop_loss"])/100)
    elif price_change > 0:
      if abs((trade_size*price_change)/100)<abs((trade_size*cell["take_profit"])/100):
        profit_loss = (trade_size*price_change)/100
      elif abs((trade_size*price_change)/100)>abs((trade_size*cell["take_profit"])/100):
        profit_loss = (trade_size*cell["take_profit"])/100
    capital += profit_loss
  final = capital - 1000
  return final

def crossover(keys):
    pair = random.sample(keys, 2)
    num1, num2 = pair[0], pair[1]
    keys.remove(num1)
    keys.remove(num2)
    num3, num4 = keys[0], keys[1]
    split_point = random.randint(1, 5)

    new_num1 = num1[:split_point] + num2[split_point:]
    new_num2 = num2[:split_point] + num1[split_point:]
    new_num3 = num3[:split_point] + num4[split_point:]
    new_num4 = num4[:split_point] + num3[split_point:]
    return [new_num1, new_num2, new_num3, new_num4]

def mutate_numbers(numbers, mutation_rate=0.05):
    mutated_numbers = []

    for num in numbers:
        if random.random() < mutation_rate:
            index = random.randint(0, len(num) - 1)
            new_digit = str(random.randint(0, 9))
            num = num[:index] + new_digit + num[index+1:]
        mutated_numbers.append(num)

    return mutated_numbers

def revert(cells):
    reversed_population = []

    for key, state in cells.items():
        item = {"stop_loss": 0, "take_profit": 0, "trade_size": 0}

        stop_loss_condition = state[0]
        take_profit_condition = state[1]
        trade_size_condition = state[2]

        if stop_loss_condition == '1':
            item["stop_loss"] = float(key[0:2]) / 10
        elif stop_loss_condition == '0':
            item["stop_loss"] = int(key[0:2])

        if take_profit_condition == '1':
            item["take_profit"] = float(key[2:4]) / 10
        elif take_profit_condition == '0':
            item["take_profit"] = int(key[2:4])

        if trade_size_condition == '1':
            item["trade_size"] = float(key[4:6]) / 10
        elif trade_size_condition == '0':
            item["trade_size"] = int(key[4:6])

        reversed_population.append(item)

    return reversed_population

def populationtodict(population, states):
  cells = {}
  state = None
  value = 0
  stval = ""
  for i in population:
    if isinstance(i["stop_loss"], float):
      state = "1"
      i["stop_loss"] = int(i["stop_loss"]*10)
    else:
      state = "0"
    if isinstance(i["take_profit"], float):
      state += "1"
      i["take_profit"] = int(i["take_profit"]*10)
    else:
      state += "0"
    if isinstance(i["trade_size"], float):
      state += "1"
      i["trade_size"] = int(i["trade_size"]*10)
    else:
      state += "0"
    value += i["trade_size"]
    value += i["take_profit"]*100
    value += i["stop_loss"]*10000
    value += 1000000
    value = str(value)
    for i in range(1, len(value)):
      stval += value[i]
    cells[stval] = state
    states.update(cells)
    state = 0
    value = 0
    stval = ""
  return cells

def lab_task_1(capital, price, population, generations, final_profit, states):
  for i in range(generations+1):
  ####################################################
    profits = []
    for i in range(len(population)):
      arr = population[i]
      d1 = round(fitness(arr, capital, price),2)
      profits.append(d1)
  ####################################################
    cells = populationtodict(population, states)
  ####################################################
    keys = list(cells.keys())
    for i in range(len(profits)):
      final_profit[keys[i]]=profits[i]
  ####################################################
    keys = crossover(keys)
    mutated_list = mutate_numbers(keys)
    val = list(cells.values())
    newcells = {}
    for i in range(len(mutated_list)):
      newcells[mutated_list[i]] = val[i]
  ####################################################
    population = revert(newcells)
    cells = newcells
  ####################################################
  return final_profit

output =  lab_task_1(capital, price, population, generations, final_profit, states)
maxkey = max(output, key=output.get)
maxvalue = output[maxkey]
state = states[maxkey]
dict1 = {str(maxkey) : str(state)}
dict1 = revert(dict1)
maxdict = dict1[0]
print(f"\"best_strategy\": {maxdict}, \"final_profit\": {maxvalue}")
```

```python
import random
import math

def strength(x):
  return math.log2(x+1)+(x/10)
def utility(maxV, minV):
  return (strength(maxV)-strength(minV)+(((-1)**(random.choice([0, 1])))*(random.randint(1, 10)/10)))
def minimax(depth, is_max, alpha, beta, maxV, minV):
    if depth == 5:
        return utility(maxV, minV)
    if is_max:
        v = float('-inf')
        for i in range(2):
            ma = minimax(depth + 1, False, alpha, beta, maxV, minV)
            v = max(v, ma)
            alpha = max(alpha, v)
            if beta <= alpha:
                break  # Beta cutoff
        return v
    else:
        v = float('inf')
        for i in range(2):
            mi = minimax(depth + 1, True, alpha, beta, maxV, minV)
            v = min(v, mi)
            beta = min(beta, v)
            if beta <= alpha:
                break  # Alpha cutoff
        return v
def game(maxv, minv, startplayer):
  result = minimax(0, True, float('-inf'), float('inf'), maxv, minv)
  if startplayer == 0:
    if result < 0:
      winner = "Magnus Carlsen"
    elif result == 0:
      winner = "Draw"
    else:
      winner = "Fabiano Caruana"
  else:
    if result < 0:
      winner = "Fabiano Caruana"
    elif result == 0:
      winner = "Draw"
    else:
      winner = "Magnus Carlsen"
  return winner, result

startplayer = int(input("Enter starting player for game 1 (0 for Carlsen, 1 for Caruana): "))
carlsen = int(input("Enter base strength for Carlsen: "))
caruana = int(input("Enter base strength for Caruana: "))
final = {"Magnus Carlsen": 0, "Fabiano Caruana": 0, "Draw": 0}
for i in range(4):
  if i % 2 == 0:
    maxplayer= 0
    maxv = carlsen
    minv = caruana
  else:
    maxplayer = 1
    maxv = caruana
    minv = carlsen
  winner, result = game(maxv, minv, maxplayer)
  final[winner] += 1
  print(f"Game {i + 1} Winner: {winner} (Utility value: {result:.2f})")
print("\nOverall Results:")
print(f"Magnus Carlsen Wins: {final['Magnus Carlsen']}")
print(f"Fabiano Caruana Wins: {final['Fabiano Caruana']}")
print(f"Draws: {final['Draw']}")

if final["Magnus Carlsen"] > final["Fabiano Caruana"]:
  print("Overall Winner: Magnus Carlsen")
elif final["Fabiano Caruana"] > final["Magnus Carlsen"]:
  print("Overall Winner: Fabiano Caruana")
else:
  print("Overall Winner: Draw")
```
