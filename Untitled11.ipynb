```python
input = open('/content/input', mode = 'r')
output = open('/content/output', mode = 'w')
fl = input.readline()
fl = fl.split(' ')
n = int(fl[0])
m = int(fl[1])

adjMatrix = []
for i in range(n+1):
  adjMatrix.append([0 for i in range(n+1)])

for i in range(m):
  fl = input.readline()
  fl = fl.split(' ')
  adjMatrix[int(fl[0])][int(fl[1])] = int(fl[2])

for i in range(n+1):
  for j in range(n+1):
    if j != n:
      print(adjMatrix[i][j], end = ' ', file = output)
    else:
      print(adjMatrix[i][j], end = '\n', file = output)

output.close()
```

```python
input = open('/content/input', mode = 'r')
output = open('/content/output1', mode = 'w')
fl = input.readline()
fl = fl.split(' ')
n = int(fl[0])
m = int(fl[1])
adjMatrix = []
for i in range(n+1):
  adjMatrix.append([0 for i in range(n+1)])

for i in range(m):
  fl = input.readline()
  fl = fl.split(' ')
  adjMatrix[int(fl[0])][int(fl[1])] = int(fl[2])

for i in range(n+1):
  print(f'{i} : ', end = '', file = output)
  for j in range(n+1):
    if adjMatrix[i][j] != 0 and  j != n:
      print(f'({j},{adjMatrix[i][j]})', end = ' ', file = output)
    elif adjMatrix[i][j] != 0 and  j == n:
      print(f'({j},{adjMatrix[i][j]})', end = '\n', file = output)
    elif adjMatrix[i][j] == 0 and  j == n:
      print("\n", file = output)
output.close()
```
