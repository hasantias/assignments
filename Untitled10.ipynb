```python
def deleteusingpredecessor(root,val):
  if root == None:
      return None
  if root.elem == val:
      #4possibility---1. no child
                      #2. left or right side child
                      #3. both site child
      #no child
      if root.left == None and root.right ==None:
          return None
      #right site child
      if root.left ==None and root.right !=None:
          return root.right
      # left site child
      if root.left!=None and root.right == None:
          return root.left
      # both site child
      if root.left!=None and root.right!=None:
          temp = root.left
          while temp.right!=None:
              temp = temp.right 
          root.elem = temp.elem
          root.left = deleteusingpredecessor(root.left,root.elem)

  elif root.elem>val:
      root.left = deleteusingpredecessor(root.left,val)
  else:
      root.right = deleteusingpredecessor(root.right,val)
  return root
  

def deleteusingsuccessor(root,val):
  if root == None:
      return None
  if root.elem == val:
      #4possibility---1. no child
                      #2. left or right side child
                      #3. both site child
      #no child
      if root.left == None and root.right ==None:
          return None
      #right site child
      if root.left ==None and root.right !=None:
          return root.right
      # left site child
      if root.left!=None and root.right == None:
          return root.left
      # both site child
      if root.left!=None and root.right!=None:
          temp = root.right
          while temp.left!=None:
              temp = temp.left 
          root.elem = temp.elem
          root.right = deleteusingsuccessor(root.right,root.elem)

  elif root.elem>val:
      root.left = deleteusingsuccessor(root.left,val)
  else:
      root.right = deleteusingsuccessor(root.right,val)
  return root
```

```python
def hermonicSum(n):
  if n>0:
    return 1/n + hermonicSum(n-1)
  else:
    return 0
hermonicSum(7)
```
